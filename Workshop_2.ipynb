##Perceptron(Neuron), Cost Function, Optimizer

Keras provides us with many different fundamental types of deep learning techniques, using which we can make different type of models and architectures.
<br><br>
!['Perceptron'](https://miro.medium.com/max/645/0*LJBO8UbtzK_SKMog)<br><br>
A Neuron in Neural Networks is just a fancy name for (a*x) + b

In a single Neuron, we have 'a','x' and 'b' as just scalars and as we keep increasing feature dimensions and number of neurons in a layer, shapes of 'a','x' and 'b' changes.

<br><br>
In keras, we make a model by adding a layer at a time, stacking them and making a Data-Flow graph which will be used for back-propagation.
<br><br>
A layer in the DL model is a set of operations which will be performed on the data. A sequence of layer makes a sequence of operations and makes a complex non-linear function which we call a Deep Learning Model
<br><br>
We define a model using 'Sequential' class in keras.

We add Neurons in a model using 'Dense' class in keras.
<br><br>

Let's make a model for y = 6*x + 3

```
#creating dataset
#for simple function like this, we need very little data

x = np.array(list(range(1,11)))
y = (np.array(list(range(1,11))) * 6) + 3

x = x.reshape(10,1)
y = y.reshape(10,1)

print('X:',x)
print('Y:',y)
```

```
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(1,input_shape=[1]))
model.summary() #prints summary (layer-name,output shape-layer wise,parameters-layer wise) and total parameters in a model
#parameters: real numbers with which some data will interact with, in model (a*x + b), a,b are parameters, also called weights and biases
```

###Cost/Loss Function

Cost Function
  : The mathematical function which will calculate the distance between prediction done by model and the actual label
<br><br>
We have many types of mathematical functions to calculate the distance between two points. "Mean-Squared-Error", "Mean-absolute-Error" etc
<br><br>
There is a new type of such function you will learn called "Binary log-loss" which is used when we do classification as we have labels in form of '0's and '1's

In classification, model outputs a number between 0-1 and we compare it using this formula:
!['Binary log-loss'](https://i.stack.imgur.com/0rA2S.png)

```
# we will use mean-squared-error in this example as we have real-numbers as output of the model

loss = tf.keras.losses.MeanSquaredError()
```

###Optimizers

Optimizers are the functions that reduces the distance between predicted value of the model and the actual label by changing the parameters of the model
<br><br>
There are different types of optimizers, like SGD,RMSprop,ADAM etc
These are complex functions that changes the weights and biases of the model using gradient descent based on the loss calculated by loss function.

```
#we will use SGD optimizer here because it's simple as our model is also simple
optimizer = tf.keras.optimizers.SGD()
```

```
#now we will set/compile our model on the chosen loss function and optimizer
model.compile(loss=loss,optimizer=optimizer)
```

```
#now we will do training with the small dataset that we created
history = model.fit(x,y,epochs=1000,batch_size=5)
```

###Artificial Neural Networks

Practial Implementation of ANNs requires a dataset, so we will use 'mnist' dataset, a task in which computer has to determine which handwritten digit is in the image. We have 10 digits [0-9], and 60k images of these 10 digits for training and 10k images for testing the model.

This is a classification problem as we are determining a digit from [0-9] in an image, so there are some changes that we will be required to make in labels of the dataset for ANNs to work.

Let's first load the dataset

```
(x_train, y_train), (x_test,y_test) = tf.keras.datasets.mnist.load_data()

#x_train (features used for training)
#y_train (labels for the respective features used for training)

#x_test (features used for testing)
#y_test (labels for the respective features used for testing)
print('\nDimensions of features and labels')
print('X train',x_train.shape)
print('Y train',y_train.shape)
print('X test',x_test.shape)
print('Y test',y_test.shape)
```

Let's see how features and labels look like

```
single_feature_image = x_train[10000]
label_of_image = y_train[10000]

print('Number in the image:',label_of_image,'\nImage:')
plt.imshow(single_feature_image)
```

Now, let's make a ANN for this task.

We have an image of shape (28 x 28), but ANNs don't take matrices or matrix as an input, it only takes a vector as one layer can only have n-Neurons, not (n x m) neurons, so we need to reshape the image matrix into a vector, and that vector will be of shape (28 x 28) = 784

```
model = tf.keras.models.Sequential()

model.add(tf.keras.layers.Dense(1024,input_shape=(784,),activation='sigmoid'))
model.add(tf.keras.layers.Dense(512,activation='sigmoid'))
model.add(tf.keras.layers.Dense(256,activation='sigmoid'))
model.add(tf.keras.layers.Dense(128,activation='sigmoid'))
model.add(tf.keras.layers.Dense(10,activation='softmax'))

model.summary()
```

Why output of the model is a 10-dimensional vector?

Let's say that output of the model is a number from [0-9]. We gave image of digit '9' as input to the model and it predicted '4', so how much distance is there between prediction and actual label, we can't say that the distance/loss is '9-4=5' because we are doing classification, we can just say that model is wrong for not predicting correctly.
<br><br>
Here model predicting '6' instead of '4' or predicting '9' instead of '4' has the same meaning that its prediction was wrong, we cannot say that predicting '6' was closer to '4' then predicting '9'. Hence we have a new mechanism for labels and output of the model, so that we can calculate loss for all wrong predictions equally.
<br><br>
One-Hot encoding
  : In digital circuits and machine learning, a one-hot is a group of bits among which the legal combinations of values are only those with a single high (1) bit and all the others low (0).

Above mentioned is 'Wikipedia' definition, let's make it simple.

It just means that we take a list of 10 numbers and attach a label to each number. And if the any number in the list is '1' that means the actual label of that image is the label attached to that number.

Let's see it practically

```
classes = 10
y_train_one_hot = tf.keras.utils.to_categorical(y_train,10)

print('Y:',y_train[0])
print('Labels attached to each number in the list:',list(range(10)))
print('Y one-hot                                 :',y_train_one_hot[0])
print('\n\n')
print('Y:',y_train[10000])
print('Labels attached to each number in the list:',list(range(10)))
print('Y one-hot                                 :',y_train_one_hot[10000])
```

Now if model predicts '9' or '6' instead of '4', we can calculate the distance between prediction and actual label by comparing these 10 numbers in prediction and label, and distance between prediction '9' and label '4' is same as prediction '6' and label '4'. 

```
y_test_one_hot = tf.keras.utils.to_categorical(y_test,10)
```

Now we will reshape the input matrix into a vector

```
x_train_vector = x_train.reshape(60000,784)
x_test_vector = x_test.reshape(10000,784)

print('\nDimensions of features and labels')
print('X train',x_train_vector.shape)
print('Y train',y_train_one_hot.shape)
print('X test',x_test_vector.shape)
print('Y test',y_test_one_hot.shape)
```

We must first set/compile on chosen loss and optimizer

```
model.compile(optimizer="sgd", loss='categorical_crossentropy', metrics=['accuracy'])
```

now we will call model.fit method and start training

```
history = model.fit(x_train_vector,y_train_one_hot,batch_size=128,epochs=20,verbose=1)
```

```
test_evaluation = model.evaluate(x_test_vector,y_test_one_hot)
```

###Convolutional Neural Networks

Convolutional Neural Network is a technique which extracts features from images and is built on top of an ANN and combination gives us incredible results

Used mostly for images and video related tasks.

Consists of:
- Convolution
- Pooling
- Flatten
- ANNs

Convolution
  : In mathematics (in particular, functional analysis), convolution is a mathematical operation on two functions (f and g) that produces a third function (f*g) that expresses how the shape of one is modified by the other.

Above mentioned definition is taken 'Wikipedia', let's make it simple.

!['Convolution'](https://embarc.org/embarc_mli/doc/build/html/_images/image104.jpg)

<br><br>

Let's see what convolution does practically

```
kernel_matrix = np.expand_dims(np.array([[-1,0,1]]*3),axis=2)
filters = np.expand_dims(np.concatenate((kernel_matrix,kernel_matrix,kernel_matrix),axis=2),axis=3)
print('Kernel matrix:\n',kernel_matrix[:,:,0])
print('\nActual Dimensions of filters:',filters.shape)
```

```
image_input = tf.constant(np.expand_dims(image/255.,axis=0))
print('Dimensions of the image:',image_input.shape)
```

```
convolution_output = tf.nn.conv2d(image_input,filters,strides=[1,1,1,1],padding='VALID')
```

```
print('Dimensions of Convolution output:',convolution_output.shape)
```

###Pooling

Pooling is just a way to reduce dimensions of an image and mainting the co-relation between adjacent pixels

Let's see what is does

```
pooling_output = tf.nn.max_pool2d(image_input, [2,2], strides=[1,2,2,1], padding='VALID')
print('Dimensions of the original input image:',image_input.shape)
print('Dimensions of Pooling Output:',pooling_output.shape)
plt.imshow(pooling_output.numpy()[0,:,:,:])
```

Now these poolings and Convolutions operations are performed one after another and that does the process of feature extraction from the image, and the feature extraction combined with ANNs makes CNNs and gives the results we desire

Let's see how to make CNNs in Keras, We will use 'MNIST' dataset

```
CNN = tf.keras.models.Sequential(name='LeNet')

CNN.add(tf.keras.layers.Conv2D(filters=6,kernel_size=(5,5),strides=(1,1),padding='valid',activation='tanh',input_shape=(32,32,1)))
CNN.add(tf.keras.layers.AveragePooling2D(pool_size=(2,2),strides=(2,2)))
CNN.add(tf.keras.layers.Conv2D(filters=16,kernel_size=(5,5),strides=(1,1),padding='valid',activation='tanh'))
CNN.add(tf.keras.layers.AveragePooling2D(pool_size=(2,2),strides=(2,2)))
CNN.add(tf.keras.layers.Conv2D(filters=120,kernel_size=(5,5),strides=(1,1),padding='valid',activation='tanh'))
CNN.add(tf.keras.layers.Flatten())
CNN.add(tf.keras.layers.Dense(units=84,activation='tanh'))
CNN.add(tf.keras.layers.Dense(units=10,activation='softmax'))

CNN.summary()
```

Resizing images to the input dimension of the CNN

```
def increase_size(image):
    return cv2.resize(image,(32,32),interpolation=cv2.INTER_CUBIC)

x_train_32 = []
x_test_32 = []

for i in range(x_train.shape[0]):
    x_train_32.append(increase_size(x_train[i]))

x_train_32 = np.array(x_train_32).reshape(60000,32,32,1)


for i in range(x_test.shape[0]):
    x_test_32.append(increase_size(x_test[i]))

x_test_32 = np.array(x_test_32).reshape(10000,32,32,1)


print('New Dimensions with resized images:')
print('X train',x_train_32.shape)
print('Y train',y_train_one_hot.shape)
print('X test',x_test_32.shape)
print('Y test',y_test_one_hot.shape)
```

Let's set/compile CNN on loss and optimizer

```
CNN.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```
history = CNN.fit(x_train_32,y_train_one_hot,batch_size=64,epochs=20,verbose=1)
```

```
evaluation = CNN.evaluate(x_test_32,y_test_one_hot)
```
